# Object detection with OpenCV DNN


OpenCV provides a `dnn` bundle that comes with serveral methods of reading pre-trained models from `
Darknet`, `tensorflow`, or `caffe` networks for ojbect detection.

This micro-package provides you a wrap-up to read any of those models and use it for prediction.

You can choose one of the models provided here:
- [Tensorflow 2](https://github.com/tensorflow/models/blob/master/research/object_detection/g3doc/tf2_detection_zoo.md)
- [Darknet (yolo)](https://github.com/AlexeyAB/darknet/wiki/YOLOv4-model-zoo)
- [Caffe](https://modelzoo.co/framework/caffe)

#### Function:
- `forward()` : feed the image through loaded DNN to detect objects
- `post_process()` : extract class ids of objects, its score and bounding box
- `draw_()` : draw selected classes
- `draw_all()` : draw all classes

```python
image = cv2.imread('imgs/demo.png')

outs_ = HD.forward(image)
width_ = image.shape[1]
height_ = image.shape[0]
class_ids_, confidences_, boxes_ = HD.post_process(outs_, width_, height_)

image_2 = HD.draw_(image.copy(), class_ids_, confidences_, boxes_)
```

```python
plt.figure(figsize = (15,12))
plt.imshow(image_2, interpolation='nearest')
```

```python
# Load Object Detection Model
f_loc = 'model-zoo/darknet/'
f_name = 'yolov4'
HD = ObjectDetector(f_loc + f_name +'.weights', f_loc + f_name +'.cfg', f_loc +'coco.names')

vc = cv2.VideoCapture("videos/demo.mp4")

# Define the codec and create VideoWriter object
width = int(vc.get(cv2.CAP_PROP_FRAME_WIDTH) + 0.5)
height = int(vc.get(cv2.CAP_PROP_FRAME_HEIGHT) + 0.5)

width_ = 600
height_ = 600
size = (width_, height_)

fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter('videos/'+ f_name +'_output_croped_w_location.mp4', fourcc, 20.0, size)

# Check if video/camera opened successfully
if (vc.isOpened()== False):
    print("Error opening video stream or file")
else:
    while (vc.isOpened()):
        (grabbed, frame) = vc.read()
        if not grabbed:
            print("Can't receive frame (stream end?). Exiting ...")
            break

        # Object detection
        start = time.time()
        outs_ = HD.forward(frame)
        class_ids_, confidences_, boxes_ = HD.post_process(outs_, width, height)
        end = time.time()


        start_drawing = time.time()
        frame_ = HD.draw_(frame.copy(), class_ids_, confidences_, boxes_)
        end_drawing = time.time()

        # FPS
        fps_label = "FPS: %.2f" % (1 / (end - start))
        cv2.putText(frame_, fps_label, (0, 25), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 0), 2)

        out.write(frame_)
        cv2.imshow("detections", frame)
        cv2.imshow("localization", frame_)
        if cv2.waitKey(1) == ord('q'):
            break

# Release everything if job is finished
vc.release()
out.release()
cv2.destroyAllWindows()
```
